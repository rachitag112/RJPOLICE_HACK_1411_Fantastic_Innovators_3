```python
import keras_cv

# 创建Yolo模型加载预训练参数
yolo_model = keras_cv.models.YOLOV8Detector.from_preset('yolo_v8_m_pascalvoc', bounding_box_format='xywh')
```

```python
# 加载预测图像
import keras
import numpy as np

filepath = keras.utils.get_file(origin="https://i.imgur.com/gCNcJJI.jpg")
image = keras.utils.load_img(filepath)
image = np.array(image, ndmin=4)
# 调整图像大小
inference_resize = keras_cv.layers.Resizing(
    640, 640, pad_to_aspect_ratio=True, bounding_box_format='xywh'
)

image_batch = inference_resize(image)
```

```python
# 模型推理
result = yolo_model.predict(image_batch)
```

```python
# 绘制预测结果
from keras_cv import visualization

visualization.plot_bounding_box_gallery(
    image_batch,
    value_range=(0,255),
    rows=1,
    cols=1,
    y_pred=result,
    scale=5,
    font_scale=0.7,
    bounding_box_format='xywh'
)
```
